### 1. Introduction 

Citigroup Inc. is an investment bank and financial services corportation headquarted in Lower Manhattan, New York. It was founded in 1998, and has grown to become the third largest bank in the United States. <br>

In this project, we study employment at Citigroup. Our dataset contains labor data based on region, role, gender, time period, number of employees, and inflow/outflow. We begin our report studying general trends on employment before examining these closer in the context gender. We end with an in-depth of analysis of these trends in the Arab World.

***

### 2. Variables

- <b>Company</b>: Name of the company 
- <b>Region</b>: Region name
- <b>Seniority</b>: The level of seniority of a position, where there are 4 possible values: 1, 2, 3, and 4. The most junior level is 1 and the most senior level is 4.
- <b>Gender</b>: The estimated gender, based on the first names of employees. The probabilities for names Are derived from the US social security administration. 
- <b>Month</b>: The month and year of the position. Headcount uses whether an employee was working in that position at any point during the month. Inflows and outflows use the month of the employeeâ€™s start date and end date, respectively.
- <b>Count</b>: Expectation of the number of employees in the position represented in that row at that date
- <b>Inflow</b>: Expectation of the number of employees moving into position specified at that month
- <b>Outflow</b>: Expectation of the number of employees moving out of position specified at that month

Reference: Revelio Labs Data Dictionary

#### 5.2 Average Increase Per Year

Is the number of female or male employees increasing faster? In order to quantify this, we can fit straight lines to both and examine their slopes. 

```python
from sklearn import linear_model
from scipy.stats import linregress

fig,ax = plt.subplots()
regr = linear_model.LinearRegression()

increases = []
ax = plt.gca()
for i in Gender['gender'].unique():
    
    df = Gender.loc[Gender['gender']==i,:].groupby(["Year", "gender"])[['count']].sum().reset_index()
    df['Year']= df['Year'].astype(int)
    
    x = np.array(df['Year'].tolist()).reshape((-1,1))
    y = np.array(df['count'].tolist())
   
    regr.fit(x, y)
    ax.plot(x, regr.predict(x), color='black', linewidth=1)
    
    xx = df['Year'].tolist()
    yy = df['count'].tolist()
    slope, intercept, r_value, p_value, std_err = linregress(xx, yy)
    increases.append(slope)
    
    df.plot(x='Year', y='count',kind='scatter',ax=ax)
    
ax.set_title('Total Employee Count Over the Years by Gender',size=15)
ax.set_ylabel('Employee Count',size=14)
ax.set_xlabel('Year',size=14)
ax.text(2019, 1200000, 'Average female increase per year â‰ˆ 6111')
ax.text(2019, 1500000, 'Average male increase per year â‰ˆ 8297')

#NEED TO FIX X AXIS TO SHOW EACH YEAR SEPRATELY
#take log - why = 
```

### 6. Gender Ratio: A Closer Look at Trends in Female Employment

#### How do the above trends in gender differ by region, seniority, and role?

6.1 Region <br>

Which regions show the best and worst prospects for female employment by Citigroup? Let us consider 2016, the best year for female employment according to the previous graphs.

```python
fig,ax=plt.subplots()
A = Gender.loc[(Gender['gender']=='female')&(Gender['Year']=='2016')].groupby('region')['count'].sum()
A.plot(ax=ax, kind='bar', color='black', alpha=0.6)
ax.get_children()[list(A.index).index("Northern America")].set_color("g")
ax.get_children()[list(A.index).index("Northern Africa")].set_color("r")
ax.set_title("Female Employment By Region (2016)")
ax.set_ylabel("Female Employee Count")
ax.set_xlabel("Region")

#note: need to remove the last bar called 'empty'
#this is for all the years. 
```

Northern America shows the highest female employment for Citigroup, and at the other end of the spectrum is Northern Africa. <br>

However, this visualization is misleading. It is highly plausible that Northern America shows the highest female employment because Citigroup is an American company headquartered in the US and subsequently employs a larger workforce in and around the region. Simply employee count, hence, is not an accurate indicator of gender-based discrepancies. So while we can say with certainty that North America shows the best prospects for employment by Citigroup in general, this graph says nothing about the proportion of females employed by region.

We thus re-run our experiments using a better indicator - <i>gender ratio</i> - which we calculate as follows:

\begin{equation*}
Gender Ratio = \frac{Female Employee Count}{Total Employee Count}
\end{equation*}

Once again, we are trying to answer the question - Which region showed the best and worst prospects for female employment by Citigroup in 2016?

```python
f2= Gender.loc[(Gender['gender']=='female')&(Gender['Year']=='2016'), ['gender', 'count', 'region']]
f2_counts_by_region = np.array(f2.groupby('region')[['count']].sum()['count'].to_list())

m2 = Gender.loc[(Gender['gender']=='male')&(Gender['Year']=='2016'), ['gender', 'count', 'region']]
m2_counts_by_region = np.array(m2.groupby('region')[['count']].sum()['count'].to_list())

ratios2 = f2_counts_by_region/(f2_counts_by_region + m2_counts_by_region)
regions = ['Arab States', 'Central America and the Caribbean','Central and Western Asia', 'Eastern Asia', 'Eastern Europe','Northern Africa', 'Northern America', 'Northern Europe','Pacific Islands', 'South America', 'South-Eastern Asia','Southern Asia', 'Southern Europe', 'Sub-Saharan Africa','Western Europe', 'empty']

fig,ax=plt.subplots()
ax.bar(regions,list(ratios2),color=['black','black','black','black','green','black','black','black','black',
                                    'black','black','red','black','black','black','black'],alpha=0.6)
#ax.get_children()[list(A.index).index("Northern America")].set_color("g")
#ax.get_children()[list(A.index).index("Northern Africa")].set_color("r")
plt.xticks(rotation='vertical')
ax.set_ylim(0.3,0.55)
ax.set_title('Gender Ratio By Region(2016)')
ax.set_xlabel('Year')
ax.set_ylabel('Gender Ratio')
```

Clearly, Eastern Europe shows the highest gender ratio and Southern Asia shows the lowest. This is in contrast to the previous incorrect graph (where count was plotted, instead of ratio) where Northern America had the highest number of female employees.

#### 6.2 Gender Ratio Over the Years

Now that we have introduced gender ratios, let us see how they have evolved over the years.

```python
f3 = Gender.loc[Gender['gender']=='female', ['gender', 'count', 'Year']]
f3_counts_by_year = np.array(f3.groupby('Year')[['count']].sum()['count'].to_list())

m3 = Gender.loc[Gender['gender']=='male', ['gender', 'count', 'Year']]
m3_counts_by_year = np.array(m3.groupby('Year')[['count']].sum()['count'].to_list())

ratios3 = f3_counts_by_year/(f3_counts_by_year+m3_counts_by_year)
years = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]

fig,ax=plt.subplots(figsize=(10,6))
#plt.bar(years,list(ratios3))
plt.ylim(0.425,0.45)
ax.set_xticks(np.array(years))
ax.bar(years,list(ratios3),color=['black','black','red','black','black','black','black','black','green',
                                    'black','black','black','black','black','black','black'],alpha=0.6)
plt.title('Gender Ratio Over the Years')
plt.xlabel('Year')
plt.ylabel('Gender Ratio')
plt.tight_layout()
```

2010 and 2011 were the worst years for gender ratio at Citigroup. In contrast, the years 2013 and 2016 showed the highest gender ratios, with values over 0.440. 

#### 6.3 Gender Ratio by Seniority

```python
f4= Gender.loc[(Gender['gender']=='female')&(Gender['Year']=='2016'), ['gender', 'count', 'seniority']]
f4_counts_by_seniority = np.array(f4.groupby('seniority')[['count']].sum()['count'].to_list())

m4 = Gender.loc[(Gender['gender']=='male')&(Gender['Year']=='2016'), ['gender', 'count', 'seniority']]
m4_counts_by_seniority = np.array(m4.groupby('seniority')[['count']].sum()['count'].to_list())

ratios4 = f4_counts_by_seniority/(f4_counts_by_seniority + m4_counts_by_seniority)
seniority_levels = ['1', '2', '3', '4']

fig,ax=plt.subplots(figsize = (7,5))
explode = (0.1, 0, 0, 0)
#ax.bar(seniority_levels, ratios4)
ax.pie(ratios4, labels=[1,2,3,4], shadow=True, explode=explode, autopct='%1.1f%%',startangle=90)
ax.axis('equal')
ax.get_children()[list(A.index).index("Northern America")].set_color("g")
ax.get_children()[list(A.index).index("Northern Africa")].set_color("r")
ax.set_title('Gender Ratio By Superiority(2016)', size = 15)
ax.legend()
plt.tight_layout()

```

#### 7.1 Inflow and Outflow of male vs. female employees in this region.

#### 7.2 Seniority 
Another interesting aspect to examine is the number of high-seniority positions held by men vs. women. Note the graph below:

```python
fig,ax = plt.subplots()

Seniority = Arab.loc[(Arab['seniority']==3)|(Arab['seniority']==4),['seniority','gender','count','Year']].sort_values('Year',ascending=True)
Seniority_male = Seniority.loc[Seniority['gender']=='male',].groupby('Year')[['count']].sum().plot.bar(ax=ax, figsize=(13,5))
Seniority_female = Seniority.loc[Seniority['gender']=='female',].groupby('Year')[['count']].sum().plot.bar(ax=ax,figsize=(13,5),color='red')

ax.legend(['male','female'])
ax.set_ylabel('Number of positions')
```

As we can see from the graph above, the number of positions held by male in seniority level 3 and 4 is more than double of female for past 10 years with no sign of levelling in the future.

#### 7.3 Gender Wise Hiring Rates and Attrition Rates

```python
fig,ax = plt.subplots(nrows=1, ncols=2)

HRate_F = Arab.loc[Arab['gender']=='female',:].groupby('Year')['inflow','count'].sum()
HRate_F['Rate'] = (HRate_F['inflow']/HRate_F['count'])*100
HRate_F['Rate'].plot(ax=ax[0],figsize=(13,5))

HRate_M = Arab.loc[Arab['gender']=='male',:].groupby('Year')['inflow','count'].sum()
HRate_M['Rate']= (HRate_M['inflow']/HRate_M['count'])*100
HRate_M['Rate'].plot(ax=ax[0],figsize=(13,5))

HRate_W = Arab.groupby('Year')['inflow','count'].sum()
HRate_W['Rate']= (HRate_W['inflow']/HRate_W['count'])*100
HRate_W['Rate'].plot(ax=ax[0],figsize=(13,5))




AtRate_F = Arab.loc[Arab['gender']=='female',:].groupby('Year')['outflow','count'].sum()
AtRate_F['Rate'] = ((AtRate_F['outflow']/AtRate_F['count'])*100)
AtRate_F['Rate'].plot(ax=ax[1],figsize=(13,5))

AtRate_M = Arab.loc[Arab['gender']=='male',:].groupby('Year')['outflow','count'].sum()
AtRate_M['Rate'] = ((AtRate_M['outflow']/AtRate_M['count'])*100)
AtRate_M['Rate'].plot(ax=ax[1],figsize=(13,5))

AtRate_Whole = Arab.groupby('Year')['outflow','count'].sum()
AtRate_Whole['Rate'] = ((AtRate_Whole['outflow']/AtRate_Whole['count'])*100)
AtRate_Whole['Rate'].plot(ax=ax[1],figsize=(13,5),x='Year')


ax[0].set_title('Gender-based and Overall Hiring Rates')
ax[1].set_title('Gender-based and Overall Attrition rates')
ax[0].legend(['female','male','Arab_Whole'])
ax[1].legend(['female','male','Arab_Whole'])
```

The graph on the left side shows that the hiring rate for female in recent years is in upward trend although the same for the whole Arab states is decreasing. The attrition rate for female seems to be fairly constant whereas it has been decreasing for male and the whole Arab States.
